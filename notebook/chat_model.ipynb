```python
import os 
from dotenv import load_dotenv

load_dotenv()

api_key =os.environ.get("OPENAI_API_KEY")

if not api_key:
    raise ValueError("open_api_key is not in enviroment value")

# else:
    # print(api_key)
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini")
llm.invoke("hello")
```

```python
from langchain_core.prompts import PromptTemplate

prompt=PromptTemplate(template="translate the following text to kannada:{input}")

client = prompt | llm 

client.invoke({
    "input":"i love langchain"
}).content
```
